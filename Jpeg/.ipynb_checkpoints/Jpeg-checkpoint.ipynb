```python
# RGB -> YUV

def RGB2YUV(rgb_img):
    mat = np.array([[ 0.29900, -0.16874,  0.50000],
                 [0.58700, -0.33126, -0.41869],
                 [ 0.11400, 0.50000, -0.08131]])
    yuv_img = rgb_img.dot(mat)
    yuv_img += np.array([[[0, 128.0, 128.0]]])
    
    return yuv_img


# YUV -> RGB

def YUV2RGB(yuv_img):
    mat = np.array([[1.0, 1.0, 1.0],
                 [-0.000007154783816076815, -0.3441331386566162, 1.7720025777816772],
                 [1.4019975662231445, -0.7141380310058594 , 0.00001542569043522235]])
    
    rgb_img = yuv_img.dot(mat)
    rgb_img += np.array([[[-179.455, 135.459, -226.818]]])
 
    return np.cast[np.uint8](rgb_img)

yuv_img = RGB2YUV(img)

rgb_img = YUV2RGB(yuv_img)
```

```python
# Test for RGB2YUV and YUV2RGB

figure, ax = plt.subplots(1, 2, dpi=100)
ax[0].imshow(np.cast[np.uint8](np.around(rgb_img)))
ax[0].set_title("RGB->YUV->RGB Img", fontsize=10)
ax[0].set_yticks([])
ax[1].imshow(img)
ax[1].set_title("Original Img", fontsize=10)
ax[1].set_yticks([])
figure.subplots_adjust(hspace=1)
plt.show()
```

```python
# Quantify Matrix
Y_round_matrix = np.array(
    [
        [16,11,10,16,24,40,51,61],
        [12,12,14,19,26,58,60,55],
        [14,13,16,24,40,57,69,56],
        [14,17,22,29,51,87,80,62],
        [18,22,37,56,68,109,103,77],
        [24,35,55,64,81,104,113,92],
        [49,64,78,87,103,121,120,101],
        [72,92,95,98,112,100,103,99]
    ]
)

UV_round_matrix = np.array([
    [17,18,24,47,99,99,99,99],
    [18,21,26,66,99,99,99,99],
    [24,26,56,99,99,99,99,99],
    [47,66,99,99,99,99,99,99],
    [99,99,99,99,99,99,99,99],
    [99,99,99,99,99,99,99,99],
    [99,99,99,99,99,99,99,99],
    [99,99,99,99,99,99,99,99]
])
```

```python
# DCT (8 x 8 block)

def _DCT(N):
    def inner(img):
        dct_img = np.zeros(img.shape)
        A = np.zeros((N, N))
        co1 = 1 / np.sqrt(N)
        co2 = 1.41421 / np.sqrt(N)
        for i in range(N):
            for j in range(N):
                co = co1 if i == 0 else co2
                A[i, j] = co * np.cos((2*j+1)*np.pi*i/(2*N)) 
                    
        for i in range(0, img.shape[0]/N):
            for j in range(0, img.shape[1]/N):
                for k in range(3):
                    tmp = A.dot(img[i*N:(i+1)*N, j*N:(j+1)*N, k]).dot(A.T) / (Y_round_matrix if k == 0 else UV_round_matrix)
                    dct_img[i*N:(i+1)*N, j*N:(j+1)*N, k] = np.around(tmp)
                    
        return dct_img
    return inner

DCT = _DCT(8)

dct_yuv_img = DCT(yuv_img)
```

```python
# Inverse DCT

def _iDCT(N):
    def inner(img):
        idct_img = np.zeros(img.shape)
        A = np.zeros((N, N))
        co1 = 1 / np.sqrt(N)
        co2 = 1.41421 / np.sqrt(N)
        for i in range(N):
            for j in range(N):
                co = co1 if i == 0 else co2
                A[i, j] = co * np.cos((2*j+1)*np.pi*i/(2*N)) 
                    
        for i in range(0, img.shape[0]/N):
            for j in range(0, img.shape[1]/N):
                for k in range(3):
                    tmp = img[i*N:(i+1)*N, j*N:(j+1)*N, k] * (Y_round_matrix if k == 0 else UV_round_matrix)
                    idct_img[i*N:(i+1)*N, j*N:(j+1)*N, k] = A.T.dot(tmp).dot(A) 
        return idct_img
    return inner

iDCT = _iDCT(8)

re_yuv_img = iDCT(dct_yuv_img)
```

```python
# Test for Recovered Jpeg
re_rgb_img = YUV2RGB(re_yuv_img)
figure, ax = plt.subplots(1, 2, dpi=1000)
ax[0].imshow(np.cast[np.uint8](np.around(re_rgb_img)))
ax[0].set_title("Recovered Jpeg Img", fontsize=10)
ax[0].set_yticks([])
ax[1].imshow(img)
ax[1].set_title("Original Img", fontsize=10)
ax[1].set_yticks([])
figure.subplots_adjust(hspace=1)
plt.show()
```

```python
# Partial Jpeg Compression
def partial_Jpeg(img):
    yuv_img = RGB2YUV(img)
    DCT = _DCT(8)
    dct_yuv_img = DCT(yuv_img)
    return dct_yuv_img

partial_jpeg_img = partial_Jpeg(img)
```

```python
# Zigzag Conversion
def zigzag(mat):
    n = len(mat)
    lst = []
    for i in range(2*n-1):
        j_range = range(max(0, i - n + 1), min(n, i+1))
        if i % 2 == 0: j_range = reversed(j_range)
        for j in j_range:
            lst.append(mat[j][i-j])
    return np.asarray(lst)

def inverse_zigzag(lst):
    n2 = len(lst)
    n = int(np.sqrt(n2))
    assert n * n == n2, "Not valid length (Should be n = k^2, k \in Z)"
    mat = np.zeros((n, n))
    cnt = 0
    for i in range(2*n-1):
        j_range = range(max(0, i - n + 1), min(n, i+1))
        if i % 2 == 0: j_range = reversed(j_range)
        for j in j_range:
            mat[j][i-j] = lst[cnt]
            cnt += 1
    return mat
```

```python
# RLE 

def rle_encoding(lst):
    zero_cnt = 0
    output = []
    for num in lst:
        if num == 0: zero_cnt += 1
        else:
            output.append(zero_cnt)
            output.append(min(254, num+128))
            zero_cnt = 0
    output.append(255)
    output.append(255)
    return np.asarray(output, dtype=np.uint8)

def rle_decoding(rle_lst, max_len):
    
    assert len(rle_lst) % 2 == 0, "%s, %s" % (rle_lst, len(rle_lst.shape))
    output = np.zeros(max_len)
    k = 0
    for i in range(len(rle_lst)/2):
        cnt = rle_lst[i * 2]
        num = rle_lst[i * 2 + 1]
        if cnt == num == 255: break
        k += cnt
        output[k] = num - 128
        k += 1
    return output
```

```python
def ceil_by(num, base):
    mul = num / base
    return num if mul * base == num else (mul + 1) * base

print ceil_by(371, 8)
```

```python
# Integrate all parts of algorithm


def jpeg_encoding(img):
    _m, _n = img.shape[:2]
    m = ceil_by(_m, 8)
    n = ceil_by(_n, 8)
    padded_img = np.zeros([m, n, 3], dtype=np.uint8)
    padded_img[:_m, :_n] = img
    partial = partial_Jpeg(padded_img)
    
    output = []
    output.append(_m % 256)
    _m /= 256
    output.append(_m % 256)
    output.append(_n % 256)
    _n /= 256
    output.append(_n % 256)
    
    for i in range(m / 8):
        for j in range(n / 8):
            for k in range(3):
                zigzag_lst = zigzag(partial[i*8:(i+1)*8, j*8:(j+1)*8, k])
                rle_lst = rle_encoding(zigzag_lst)
                output.extend(rle_lst)
    return np.asarray(output, dtype=np.uint8)


def jpeg_decoding(jpeg_img):
    m1, m2, n1, n2 = jpeg_img[:4]
    _m = m1 + m2 * 256
    _n = n1 + n2 * 256
    m = ceil_by(_m, 8)
    n = ceil_by(_n, 8)
    img = np.zeros((m, n, 3), dtype=np.uint8)
    p = 4
    iDCT = _iDCT(8)
    for i in range(m/8):
        for j in range(n/8):
            dct_yuv_img = np.zeros((8, 8, 3))
            for k in range(3):
                last_p = p
                while p < len(jpeg_img):
                    p += 2
                    if jpeg_img[p-1] == jpeg_img[p-2] == 255:
                        break
                
                zigzag_lst = rle_decoding(jpeg_img[last_p:p], 64)
                dct_yuv_img[:, :, k] = inverse_zigzag(zigzag_lst)
            yuv_block = iDCT(dct_yuv_img)        
            img[i*8:(i+1)*8, j*8:(j+1)*8] = YUV2RGB(yuv_block)
            
    return img[:_m, :_n]
            
                
    
```

```python
# Wrapper for file I/O

def readJpeg(filename):
    img = None
    with open(filename, 'rb') as f:
        lst = []
        next_ch = f.read(1)
        while next_ch:
            lst.append(ord(next_ch))
            next_ch = f.read(1)
        img = jpeg_decoding(np.asarray(lst, dtype=np.uint8))
    return img

def writeJpeg(img, filename):
    with open(filename, 'wb') as f:
        jpeg_img = jpeg_encoding(img)
        for byte in jpeg_img:
            f.write(byte)
    
```
